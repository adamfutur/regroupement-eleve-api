```python
n = int(input("Enter a decimal number: "))
bit=[]

while n > 0:
    bit.append(n % 2)
    n = n // 2 
bit.reverse()
print(bit)
```

```python
# binary to decimal
bit = [0, 0, 1, 0]  # example binary number
dec = 0
for i in range(len(bit)):
    dec = dec + bit[i] * (2 ** (len(bit) - 1 - i))
print(dec)
```

```python
# number of the bits necessaire to represent a number
num = int(input("Enter a decimal number: "))
count = 0
while num > 0:
    num = num // 2
    count += 1
print(count)
```
